```python
#the list prodname and the total values
types = data['prodname'].value_counts()
types.reset_index()
```

```python
types_pie = types[:16]
others_sum = 0
for i in types[16:]:
    others_sum+=1
```

```python
types_pie['others'] = others_sum
```

```python
#the highst product is kindergarten total carriculunm kit, 1 st grade total curriculum kit -doer 
plt.figure(figsize=(15,12))
plt.pie(types_pie.values, labels=types_pie.index,labeldistance=1.2,autopct='%1.2f%%', shadow=True, startangle=60)
plt.title('prodname', fontsize = 20)
plt.plot()
```

```python
#the highest product sale 143 in price $595.0 
plt.figure(figsize=(20,10))
fig = sns.countplot(data['prodprice'],palette=['Orange'])
plt.xlabel('sold or unsold')
plt.ylabel('prodname')
plt.title('sold vs unsold', fontsize=15)
plt.xticks(rotation=85)
#plt.plot()
plt.plot()

for p in fig.patches:
    fig.annotate(format(p.get_height(), '.0f',), (p.get_x() +
    p.get_width()/2., p.get_height()), ha = 'center', va = 'center',
                xytext = (0, 4), textcoords = 'offset points')
```

```python
data.groupby('status')['prodid'].count()
```

```python
#the higest product total 181 Kindergarten Total Curriculum Kit
plt.figure(figsize=(20,10))
fig = sns.countplot(data[data['prodname']!='sold']['prodname'])
plt.xlabel('prodnames')
plt.ylabel('number of prodname')
plt.title('sold carculum prodname', fontsize=15)
plt.xticks(rotation=87)
plt.plot()

for p in fig.patches:
    fig.annotate(format(p.get_height(), '.0f',), (p.get_x() +
    p.get_width()/2., p.get_height()), ha = 'center', va = 'center',
                xytext = (0, 4), textcoords = 'offset points')
```

```python
# the higest prodauct custemers from USA 
plt.figure(figsize=(20,10))
fig = sns.countplot(data[data['Country']!='sold']['Country'],hue=data['status'])
plt.xlabel('prodnames')
plt.ylabel('number of prodname')
plt.title('sold carculum prodname', fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
# califorina the most large amount product sells 
plt.figure(figsize=(20,10))
fig = sns.countplot(data[data['state']!='sold']['state'],hue=data['status'])
plt.xlabel('prodnames')
plt.ylabel('number of prodname')
plt.title('sold carculum prodname', fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
dm = data.groupby(['o_date'])['status'].count()
dm.head()
```
